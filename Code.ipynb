### Separating the Target Variable

```python
X, y = df.drop(columns=["Date","Module","Style Code","Style Description","SO","FG Reference","SO/LI Worked Hours","Worked Hours"]), df["Worked Hours"]
```

```python
X
```

```python
y
```

### Splitting into training and testing dataset

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Modeling Data

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge
from sklearn.metrics import r2_score


# Define five linear regression algorithms
models = [LinearRegression(), Ridge(), Lasso(), ElasticNet(), BayesianRidge()]

# Train each algorithm on the training set and predict the values for the testing set
accuracies = []
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = r2_score(y_test, y_pred)
    accuracies.append(accuracy)
    print(f'{model.__class__.__name__} Accuracy: {accuracy:.4f}')

# Store the accuracies in a data frame
metrics = pd.DataFrame(accuracies, columns=['Accuracy'], index=['Linear Regression', 'Ridge', 'Lasso', 'Elastic Net', 'Bayesian Ridge'])

# Visualize the accuracies on a bar graph
fig, ax = plt.subplots(figsize=(10,6))
metrics.plot(kind='bar', ax=ax)
plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
plt.title('Accuracy of Linear Regression Algorithms')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
